### Schrodinger Equation for Multi-Electron Atom

Once we move beyond Hydrogen atom which has a proton and an electron, we enter the realm of multi-electron atom, say Helium atom $\left({}^2\text{He}_4 \right)$ with 2 protons and 2 electrons. Solving the Schrodinger's equation like for the Hydrogen atom becomes impossible because of the introdution of interaction term between electrons;


From this we can write the generalized expressions for Hamiltonian for any atoms with atomic number $Z$

$$H = -\frac{ℏ^2}{2 m_e} \sum \limits _i ∇^2_i + \sum \limits _i V(r_i) + \sum \limits_{i, \, j\\ i\neq j} V(r_{ij})\, \ldots \, (1)$$

Neglecting that probelamtic term one can reduce:

$$H = -\frac{ℏ^2}{2m_e}(∇^2_1+∇^2_2) + V(r_1) + V(r_2) → H(r_1) + H(r_2)\, \ldots \, (2)$$


### Hartree Approximation

Neglecting the repulsion between electrons has a serious consequence but the Hartree approximation starts by invoking an initial ansatz that the multi-electron wavefunction can be expanded as a product of single electron wavefunctions

$$|ψ(r_1, r_2, \ldots r_i)⟩ \approx |ψ(r_1)\rangle \, |ψ(r_2)\rangle \ldots |ψ(r_i)\rangle \, \ldots \, (3)$$


------------------------------------------------------------------

Problem: Next we will discuss Slater Determinant. The ansatz in Eq. 3 that leads to Hartree method neglects one important property of electrons (fermions); The wave function of a system of fermions needs to be antisymmetric w.r.t interchanging any pair of particles.



### Slater Determinant:

Anti-symmetric wave functions can be represented as:

$$|\psi⟩(x_1, x_2) = - |\psi⟩(x_2, x_1)$$

This doesn't hold for Hartree product. As electrons are indistinguishable one natural thing to do is to construct another wavefunction by taking a linear superposition as below:

$$\begin{align} |\psi⟩(x_1, x_2) &= \frac{1}{\sqrt{2}} (|\phi\rangle_1 (x_1)\, |\phi\rangle_2 (x_2)\, - |\phi\rangle_1 (x_2)\, |\phi\rangle_2 (x_1)\,) \\ &= \frac{1}{\sqrt{2}}\begin{vmatrix} |\phi\rangle_1 (x_1) && |\phi\rangle_2 (x_1)\\ |\phi\rangle_1 (x_2) && |\phi\rangle_2 (x_2)\,    \end{vmatrix} \end{align}\, \ldots \, (4)$$

Here we can easily verify that $|\psi⟩(x_1, x_2) = - |\psi⟩(x_2, x_1)$.

If there happen to be two identical rows or two identical columns, it means that two particles share the same spatial coordinates (and spin); the determinant then equals zero, meaning that such situation is physically impossible (Pauli Exclusion principle). One can also see that the Slater determinant is anti-symmetric upon the exchange of any two electrons. We recall that if we take a matrix and interchange two of its rows, the determinant changes sign.

We usually call these $|\phi ⟩$'s as spin orbitals. It takes both the position and spin angular momentum of the particle,  so $x_i$ is actually $ (x_i, y_i, z_i, s_{z_i})$. Each column includes a different spin orbital as a function of the coordinates $(x_i)$ in each row. So, it is never possible to say that $|\phi ⟩_1$ describes the first electron and $|\phi ⟩_2$ the second electron. The terms first electron and second electron have no meaning as the electrons are indistinguishable and it is never possible to put a label on one specific electron. That's why the linear combination of them works and in the linear combination itself we see that for two electrons $|\phi ⟩_i$ appear twice.  Check [here](https://people.unipi.it/static/Cacelli/didattica/dispense.chimteo/HF_cacelli_revised.pdf) for more.  

**Note: If we do change the -'ve' sin to +'ve' we get a linear superposition that will not vanish for identical particles, these are called Bosons**     

We can simply expand the Slater determinant for multi-particle case by expanding the determinant as below:

$$\psi (x_1, x_2, \ldots x_N) = \frac{1}{\sqrt{N!}} \begin{vmatrix} |\phi\rangle_1 (x_1) && |\phi\rangle_2 (x_1) && \ldots && |\phi\rangle_N (x_1) \\ |\phi\rangle_1 (x_2) && |\phi\rangle_2 (x_2) && \ldots && |\phi\rangle_N (x_2) \\ \vdots && \vdots && \ddots && \vdots \\ |\phi\rangle_1 (x_N) && |\phi\rangle_2 (x_N) && \ldots && |\phi\rangle_N (x_N) \end{vmatrix} \equiv |\phi_1\, \phi _2\, \ldots \phi _N\rangle \, \ldots \, (5)$$

The determinant expands into a sum of permutations of orbitals, with alternating signs based on the number of permutations:

$$\psi (x_1, x_2, \ldots x_N) = \frac{1}{\sqrt{N!}} \, \sum _P (-1)^P \, \phi _{P1}(x_1)\, \phi _{P2}(x_2)\, \ldots \phi _{PN}(x_N)$$

Here, $P$ is the permulation of the indices $1, 2, \ldots N$ and $(-1)^P$ representing the sign of the permutation whether it is even of odd. Thus, the determinant inherently adds and subtracts terms based on the permutations, enforcing antisymmetry under particle exchange. If two electrons occupy the same orbital, the determinant vanishes, satisfying the Pauli exclusion principle.(chek [here](https://www.home.uni-osnabrueck.de/apostnik/Lectures/DFT-2.pdf)).

We can expand this for 3 spin orbitals as below:

$$\psi (x_1, x_2, x_3) = \frac{1}{\sqrt{3!}} \begin{vmatrix} |\phi\rangle_1 (x_1) && |\phi\rangle_2 (x_1) && |\phi\rangle_3 (x_1) \\ |\phi\rangle_1 (x_2) && |\phi\rangle_2 (x_2) && |\phi\rangle_3 (x_2) \\  |\phi\rangle_1 (x_3) && |\phi\rangle_2 (x_3) && |\phi\rangle_3 (x_3) \end{vmatrix} \equiv |\phi_1\, \phi _2\,  \phi _3\rangle \, $$

We can calculate the determinant in the traditional way as below:

$$\begin{align}\psi (x_1, x_2, x_3) &= \frac{1}{\sqrt{3!}}\left[|\phi\rangle_1 (x_1)\, \left(|\phi\rangle_2 (x_2)\, |\phi\rangle_3 (x_3) - |\phi\rangle_3 (x_2)\, |\phi\rangle_2 (x_3)\right) - |\phi\rangle_2 (x_1) \, \left(|\phi\rangle_1 (x_2) |\phi\rangle_3 (x_3) - |\phi\rangle_3 (x_2) |\phi\rangle_1 (x_3)\right)  + |\phi\rangle_3 (x_1) \left(|\phi\rangle_1 (x_2) |\phi\rangle_2 (x_3) - |\phi\rangle_2 (x_2) |\phi\rangle_1 (x_3)\right)\right] \\ &=\frac{1}{\sqrt{3!}} \left[|\phi\rangle_1 (x_1)\, |\phi\rangle_2 (x_2)\, |\phi\rangle_3 (x_3)\,- |\phi\rangle_1 (x_1)\, |\phi\rangle_3 (x_2)\, |\phi\rangle_2 (x_3)\, + |\phi\rangle_2 (x_1)\, |\phi\rangle_3 (x_2)\, |\phi\rangle_1 (x_3)\, - |\phi\rangle_2 (x_1)\, |\phi\rangle_1 (x_2)\, |\phi\rangle_3 (x_3)\, + |\phi\rangle_3 (x_1)\, |\phi\rangle_1 (x_2)\, |\phi\rangle_2 (x_3)\, - |\phi\rangle_3 (x_1)\, |\phi\rangle_2 (x_2)\, |\phi\rangle_1 (x_3)\,\right] \end{align}$$

## Second Quantization


Once we have this multi-particle wave-function, we can define creation and annihilation operator; where the creation operator $(a_i^{†})$ create a new one-particle state; For example given any arbitrary slater determinant $|\phi_1, \phi _2, \ldots, \phi _N\rangle$

$$a_i^{†} |\phi_1, \phi _2, \ldots, \phi _N\rangle = |\phi_i, \phi_1, \phi _2, \ldots, \phi _N\rangle \, \ldots (6)$$

Due to the anti-symmetric nature of the wave functions for fermions, the order in which these two creation operators act is important.

$$\begin{align} a_i^{†} \, a_j^{†}|\phi_1, \phi _2, \ldots, \phi _N\rangle = a_i^{†} |\phi _j, \phi_1, \phi _2, \ldots, \phi _N\rangle &= |\phi _i, \phi _j, \phi_1, \phi _2, \ldots, \phi _N\rangle\\ a_j^{†} \, a_i^{†}|\phi_1, \phi _2, \ldots, \phi _N\rangle = |\phi_j, \phi _i, \phi_1, \phi _2, \ldots, \phi _N\rangle &= -|\phi_i, \phi _j, \phi_1, \phi _2, \ldots, \phi _N\rangle \end{align}\, \ldots (7)$$

This automatically implies the first anti-commutator relation:

$$ a_i^{†}\, a_j^{†} + a_j^{†}\,  a_i^{†} = \{a_i^{†}, a_j^{†}\} = 0 \, \ldots (8)$$


**Note: Since for Bosons we can interchange states without any issue, instead of anti-commutation, we will get commutation relation for both creation and annihilation operators**

When $i=j$ in eq. 24, i.e. trying to create 2 states with same indices (quantum number), $a_i^{†}\, a_i^{†} = - a_i^{†}\, a_i^{†} $ we will get zero; Pauli Exclusion principle: Explicitly we can check this as below:

$$a_1^{\dagger}\, a_1^{\dagger} |\phi_2, \phi _3 \rangle = a_1^{\dagger} |\phi _1, \phi_2, \phi _3 \rangle = |\phi_1, \phi _1, \phi_2, \phi _3 \rangle = 0$$

Generalized rule:

$$a_i^{\dagger} |\phi_j, \, \ldots, \phi _z \rangle = 0\, \, \text{if}\, i\in \{j, \ldots, z\}\,  \ldots \, (9)$$


Similarly, we can define the annihilation operator as the adjoint of the creation operator $\left(a^{\dagger}\right)^{\dagger} = a$, where

$$a_i |\phi_i, \phi_j, \, \ldots, \phi _z \rangle = |\phi_j, \, \ldots, \phi _z \rangle$$

If the state is not exactly at the left then we need to interchange:

$$a_i |\phi_j, \phi_k, \phi _i \rangle = -a_i|\phi_j, \phi _i, \phi _k \rangle = a_i|\phi_i, \phi _j, \phi _k \rangle = |\phi _j, \phi _k \rangle$$


We have discussed so far about these operators operating only on 'Ket' states. Let's see what happens if we go to 'Bra' states.

$$|X\rangle = |\phi_j \phi_k\rangle = a_j^{\dagger} |\phi_k\rangle$$

This implies:  

$$\langle X| = \langle\phi_k \phi_j| = \langle \phi _k|a_j$$

As expected when we move from Ket to Bra states the Creation and Annihilation operators will behave 'opposite'.

$$\langle X|X\rangle = \langle \phi _k a_j|\phi_j \phi_k\rangle = \langle \phi _k |\phi_k\rangle = 1$$

### Anti-Comutation Rule: (Proof)

We already have a part of the anti-commutation rule for fermions in eq. 24. The genralized rule is:

$$\{a_i, a_j^{†}\} = \delta _{ij}\, \ldots (P.1.)$$

Let's try to prove it; First we consider $i=j$

$$a_i a_i^{†}|\phi _j, \, \phi _k, \, \phi _l \rangle = a_i|\phi _i, \, \phi _j, \, \phi _k, \, \phi _l \rangle = |\phi _j, \, \phi _k, \, \phi _l \rangle$$

but $$ a_i^{†} a_i|\phi _j, \, \phi _k, \, \phi _l \rangle = 0$$

as we cannot annihilate a state that does not exist. So

$\{a_i, a_i^{†}\} |\phi _j, \, \phi _k, \, \phi _l \rangle = |\phi _j, \, \phi _k, \, \phi _l \rangle \implies \{a_i, a_i^{†}\} = 1\, \ldots \, (P.2.)$

Here we assumed that state $i$ wasn't occupied; The results won't change even if we have started from a basis state $|\phi _j, \, \phi_i, \, \phi _k, \, \phi _l \rangle$




Let's now consider $i\neq j$


$$a_i a_j^{†}|\phi _i, \, \phi _k, \, \phi _l \rangle = a_i|\phi _j,\, \phi _i, \,  \phi _k, \, \phi _l \rangle = -a_i|\phi _i, \, \phi _j, \, \phi _k, \, \phi _l \rangle = - |\, \phi _j, \, \phi _k, \, \phi _l \rangle \, \ldots \, (P.3.)$$


but,


$$a_j^{†}a_i|\phi _i, \, \phi _k, \, \phi _l \rangle = a_j^{†}|\phi _k, \, \phi _l \rangle = |\phi _j, \, \phi _k, \, \phi _l \rangle$$


So, summing these two equations above:

$$a_i\,  aj^{\dagger} + a_j^{\dagger}\, a_i = 0 \implies \{a_i, a_j^{\dagger}\}=0 ; i\neq j\, \ldots \, (P.4.)$$

Combining both $\{a_i, a_i^{†}\}=1$ and $\{a_i, a_j^{†}\}=0; i\neq j$ we have the generalized anti-commutation rule Eq. P.1.

**Note: For bosons, this will be a commutation rule: $\left[a_i, a_j^{†}\right] = \delta _{ij}$**

### Occupation Number Operator $\hat{n_i} = a_i^{†} a_i $

For fermions, occupation number for a state $i$ is always either 0 or 1. It's written above with an $\hat{}$ to explicitly highlight that it is an operator.

Let's assume we have an occupation number 0 for a state $i$

$n_i |\phi _k, \phi _l, \ldots \phi _z\rangle $ i.e., $\phi _i$ does not exist; so

$a_i^{†} a_i |\phi _k, \phi _l, \ldots \phi _z\rangle $; since $\phi _i$ does not exist; annihilation operator acting on the state would return 0.

The other scenario $n_i = 1$; i.e. that state exists and it has a fermion.

$a_i^{†} a_i |\phi _i, \phi _k, \phi _l, \ldots \phi _z\rangle = a_i^{†}| \phi _k, \phi _l, \ldots \phi _z\rangle = |\phi _i, \phi _k, \phi _l, \ldots \phi _z\rangle$

What this shows that Fock states are eigenstates of the occupation number and the eigenvalues just tell us whether that state is occupied (exists) or not (does not exist).

So the total number of particles in a many-body system can be written as

$N = \sum n_i = \sum a_i ^{\dagger} a_i$

### Fock Space:


With the occupation number operator being introduced, now we can define Fock space;

In very short, we think of Fock space as an 'extension' of the single paticle Hilbert space that consists of bosonic/fermionic states of arbitrary number of particles;

Fock space for a bosonic system can have the form

$$|n_1, n_2, n_3, \ldots⟩, \, \, n_{\alpha} = 0, 1, 2, \ldots; \, \, \text{number of particles in state }\alpha$$

This is different from the Fock space of a fermionic system, which must follow Pauli Exclusion principle

$$|n_1, n_2, n_3, \ldots⟩, \, \, n_{\alpha} = 0, 1; \, \, \text{number of particles in state }\alpha$$

Total number of particles in the Fock space is always $N = \sum n_k$.

What does the phrase 'extension of Hilbert space' means? For that we need to take a look at the construction of the Fock space:

$$\mathcal{F}_{\nu} = \bigoplus _{0}^{\infty} S_{\nu}\, H^{\otimes n} = H0 ⊕ H1 \oplus \left(S_{\nu} (H\otimes H)\right) ⊕ \left(S_{\nu}\left(H\otimes H \otimes H\right)\right) ⊕ \ldots$$


Here $H0$ corresponds to the state of no particles, $H1$ the state of one particle, $S_{\nu} (H\otimes H)$ the state of two identical particles and so on...


$S_{\nu}$ is the symmetrization operator, $S_{+}$ for bosons and $S_{-}$ for for fermions. You can look back at Equation 19 to get better idea.

The operator $⨁$ in Tensor Algebra implies direct sum and if $H_1, \, H_2$ has dimensions $d_1, \, d_2$, then $H_1⨁ H_2$ will have a dimension $d_1 + d_2$.


To understand this a bit better what exactly is going on, let's consider a basis for the single-particle Hilbert space $H$ of 3 orbitals ${\phi _1, \phi _2, \phi _3}$, where these $\phi _i$'s include spatial and spin information.



Now we can construct the Fock space for fermions as below:

$$\mathcal{F} = H^{\otimes 0} \oplus H^{\otimes 1} \oplus H^{\otimes 2} \oplus H^{\otimes 3}$$

The direct sum $\oplus$ represents that 0-particle, 1-particle, 2-particle, 3-particle are independent from each other.

* The dimension of $H^{\otimes 0}$ is 1; $|000\rangle$. Vacuum state.

* The dimension of $H^{\otimes 1}$ is $3 \rightarrow ({}^3C_1)$; One particle state; We think of this as what is the space of single particle states built from these 3 orbitals:

  $|100\rangle, |010\rangle, |001\rangle$.


* The dimension of $H^{\otimes 2}$ is $3 \rightarrow ({}^3C_2)$; two particle state; We think of this as what is the space of single particle states built from these 3 orbitals:

  $|101\rangle, |110\rangle, |011\rangle$


* The dimension of $H^{\otimes 3}$ is $1 \rightarrow ({}^3C_3)$; 3 particle state, we have 3 orbitals; For Fermions we have only 1 choice

  $|111\rangle$

So total Fermionic Fock space dimension is $2^3 = 8$.
