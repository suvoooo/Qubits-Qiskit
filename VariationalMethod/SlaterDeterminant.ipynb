### Schrodinger Equation for Multi-Electron Atom

Once we move beyond Hydrogen atom which has a proton and an electron, we enter the realm of multi-electron atom, say Helium atom $\left({}^2\text{He}_4 \right)$ with 2 protons and 2 electrons. Solving the Schrodinger's equation like for the Hydrogen atom becomes impossible because of the introdution of interaction term between electrons;

Let's see this; we will have K.E terms for individual electrons, potential energy terms for each electron (interaction with positively charged nucleus), and potential energy term for the mutual repulsion of each pair of negatively charged electrons:


$$H = -\frac{ℏ^2}{2m_e}(∇^2_1+∇^2_2) + V(r_1) + V(r_2) + V(r_{12})\, \ldots \, (10)$$

where

$$V(r_1) = \frac{-e^2}{4\pi ϵ_0 r_1},\,  V(r_2) = \frac{-e^2}{4\pi ϵ_0 r_2}, \, V(r_{12}) = \frac{-e^2}{4\pi ϵ_0 |r_1-r_2|}\, \ldots \, (11)$$

From this we can write the generalized expressions for Hamiltonian for any atoms with atomic number $Z$

$$H = -\frac{ℏ^2}{2 m_e} \sum \limits _i ∇^2_i + \sum \limits _i V(r_i) + \sum \limits_{i, \, j\\ i\neq j} V(r_{ij})\, \ldots \, (12)$$

Because of the interaction term when we want to solve the Schrodinger Equation $H|\psi⟩ = E|\psi\rangle$ it becomes impossible to separate it for individual electrons; Neglecting the repulsion term we can write:

$$H = -\frac{ℏ^2}{2m_e}(∇^2_1+∇^2_2) + V(r_1) + V(r_2) → H(r_1) + H(r_2)\, \ldots \, (13)$$

so, we end up with two independent Schrödinger equations that are exactly the same as that for the hydrogen atom, except that the nuclear charge is $2e$ rather than $e$.

$$\begin{align}H(r_1)φ(r_1) &= E_1φ(r_1)\\ H(r_2)φ(r_2) &= E_2φ(r_2) \end{align} \, \ldots \, (14)$$

The fact that electrons interact through their electron-electron repulsion  means that an exact wavefunction for a multi-electron system would be a _single function that depends simultaneously upon the coordinates of all the electrons_; i.e., a multi-electron wavefunction: $|ψ(r_1, r_2, \ldots r_i)⟩$


### Hartree Approximation

The separable Hamiltonians for electrons (e.g. in Helium) once we neglect the mutual repulsion between them, tell us that the total electronic wave function would just be the product of the two Hydrogen atom wavefunctions

$$|ψ(r_1, r_2)⟩ = |ψ(r_1)\rangle \, |ψ(r_2)\rangle \, \ldots \, (15)$$

Neglecting the repulsion between electrons has a serious consequence but the Hartree approximation starts by invoking an initial ansatz that the multi-electron wavefunction can be expanded as a product of single electron wavefunctions

$$|ψ(r_1, r_2, \ldots r_i)⟩ \approx |ψ(r_1)\rangle \, |ψ(r_2)\rangle \ldots |ψ(r_i)\rangle \, \ldots \, (16)$$

With this one can write the one electron Schrodinger equations:

$$-\frac{ℏ^2}{2m_e} \, ∇^2\psi _i(r) + V(r)\, \psi_i = \epsilon _i (r) \psi _i(r)\, \ldots \, (17)$$

The potential term $V(r)$ consists of both the nuclear electron interaction and the mean field arising from $N-1$ electron.

The K.E. would be

$$\left\langle\phi_j\left|\frac{ℏ^2}{2m_e} \, ∇^2\right|\phi_j\right\rangle$$

Nuclear-electron energy (potential)

$$\left\langle\phi_j\left|\frac{-Ze^2}{|r-R|} \right|\phi_j\right\rangle$$

Electron-electron repulsive energy (potential)

$$\left\langle\phi_j(r)\, \phi_k(r')\left|\frac{e^2}{|r-r'|} \right|\phi_j(r)\phi _k(r')\right\rangle$$


Adding these all together to get the total energy $ϵ_j$ of the orbital $ϕ_j$

$$\epsilon _j = \left\langle\phi_j\left|\frac{ℏ^2}{2m_e} \, ∇^2\right|\phi_j\right\rangle + \left\langle\phi_j\left|\frac{-Ze^2}{|r-R|} \right|\phi_j\right\rangle + \left\langle\phi_j(r)\, \phi_k(r')\left|\frac{e^2}{|r-r'|} \right|\phi_j(r)\phi _k(r')\right\rangle \, \ldots \, (18)$$

Problem: Next we will discuss Slater Determinant. The ansatz in Eq. 15 that leads to Hartree method neglects one important property of electrons (fermions); The wave function of a system of fermions needs to be antisymmetric w.r.t interchanging
any pair of particles.



### Slater Determinant:

We already know that the basis of Hartree principle is to approximate wave function of many-particle system is to take product of single particle wave functions. The Hartree principle is to approximate the wave function of a many-particle system by taking products of single-particle wave functions. Electrons (fermions) are indistinguishable and a wave function representing a system of electrons must be anti-symmetric (Pauli's exclusion principle) under the exchange of any two fermions. Anti-symmetric wave functions can be represented as:

$$|\psi⟩(x_1, x_2) = - |\psi⟩(x_2, x_1)$$

This doesn't hold for Hartree product. As electrons are indistinguishable one natural thing to do is to construct another wavefunction by taking a linear superposition as below:

$$\begin{align} |\psi⟩(x_1, x_2) &= \frac{1}{\sqrt{2}} (|\phi\rangle_1 (x_1)\, |\phi\rangle_2 (x_2)\, - |\phi\rangle_1 (x_2)\, |\phi\rangle_2 (x_1)\,) \\ &= \frac{1}{\sqrt{2}}\begin{vmatrix} |\phi\rangle_1 (x_1) && |\phi\rangle_2 (x_1)\\ |\phi\rangle_1 (x_2) && |\phi\rangle_2 (x_2)\,    \end{vmatrix} \end{align}\, \ldots \, (19)$$

Here we can easily verify that $|\psi⟩(x_1, x_2) = - |\psi⟩(x_2, x_1)$.

If there happen to be two identical rows or two identical columns, it means that two particles share the same spatial coordinates (and spin); the determinant then equals zero, meaning that such situation is physically impossible (Pauli Exclusion principle). One can also see that the Slater determinant is anti-symmetric upon the exchange of any two electrons. We recall that if we take a matrix and interchange two of its rows, the determinant changes sign.

We usually call these $|\phi ⟩$'s as spin orbitals. It takes both the position and spin angular momentum of the particle,  so $x_i$ is actually $ (x_i, y_i, z_i, s_{z_i})$. Each column includes a different spin orbital as a function of the coordinates $(x_i)$ in each row. So, it is never possible to say that $|\phi ⟩_1$ describes the first electron and $|\phi ⟩_2$ the second electron. The terms first electron and second electron have no meaning as the electrons are indistinguishable and it is never possible to put a label on one specific electron. That's why the linear combination of them works and in the linear combination itself we see that for two electrons $|\phi ⟩_i$ appear twice.   

**Note: If we do change the -'ve' sin to +'ve' we get a linear superposition that will not vanish for identical particles, these are called Bosons**     

We can simply expand the Slater determinant for multi-particle case by expanding the determinant as below:

$$\psi (x_1, x_2, \ldots x_N) = \frac{1}{\sqrt{N!}} \begin{vmatrix} |\phi\rangle_1 (x_1) && |\phi\rangle_2 (x_1) && \ldots && |\phi\rangle_N (x_1) \\ |\phi\rangle_1 (x_2) && |\phi\rangle_2 (x_2) && \ldots && |\phi\rangle_N (x_2) \\ \vdots && \vdots && \ddots && \vdots \\ |\phi\rangle_1 (x_N) && |\phi\rangle_2 (x_N) && \ldots && |\phi\rangle_N (x_N) \end{vmatrix} \equiv |\phi_1, \phi _2, \ldots, \phi _N\rangle \, \ldots \, (20)$$
