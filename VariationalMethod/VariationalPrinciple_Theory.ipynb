## Approximate Methods


Exact solutions for Schrodinger's equation exist for only few ideal systems like an isolated Hydrogen Atom but, to solve more general and realistic systems we need approximation methods. We focus on _stationary states_ (corresponding to time independent Hamiltonians) and a particular approximate method; the Variational Principle.

Another approximate method 'Perturbation', depends on finding an approximate or simpler Hamiltonians that can be solved exactly but Variational method doesn't require such.  

Variational method is particularly useful for finding the ground state of a system without explicitly solving the problem. We know the definition of expectation of the energy which is as below:

$$E(\psi) = \frac{⟨\psi | H | \psi ⟩}{⟨\psi | \psi ⟩}\, \ldots\,  (1)$$

The variational principle will get an upper bound for the ground state $(E_0)$; Say we start with a trial function $|\psi ⟩$ and one can easily show that

$$E_0\leq \frac{⟨\psi | H | \psi ⟩}{⟨\psi | \psi ⟩}\, \ldots\, (2) $$

the equality condition occurs only when $|\psi\rangle$ is proportional to the true ground state $|\psi _0\rangle$.

To prove this we expand the trial function in terms of the exact eigenstates

$$|\psi\rangle = \sum \limits _n a_n |\phi _n\rangle ; \, H|\phi _n\rangle = E_n |\phi _n\rangle \, \ldots \, (3)$$

From before

$$E(\psi) = \frac{⟨\psi | H | \psi ⟩}{⟨\psi | \psi ⟩} = \frac{\sum \limits_n |a_n|^2 E_n}{\sum \limits_n |a_n|^2 }\geq \frac{E_0 \sum \limits_n |a_n|^2 }{\sum \limits_n |a_n|^2 } = E_0\, \ldots \, (4)$$


In bit more detail:

$$\begin{align}\left<\psi | H | \psi\right> &= \left(\sum _n a_n^* \langle\phi_n|\right)\, H\, \left(\sum _m a_m^* \, |\phi _m\rangle  \right) \\ &= \sum _{m, n} a_m^*\, a_n^*\, \langle\phi_n| H | \phi _m\rangle \\ &= \sum _{m, n} a_m^*\, a_n^*\, \, E_m \left\langle \phi_n|\phi_m \right\rangle \\ &=\sum _{n}\, |a_n|^2 \, E_n \end{align}\, $$


But how can we actually obtain this approximate state?

1. Start of with a 'reasonable' trial function. Trial function contains adjustable parameters $(\theta _i)$.

2. Calculate the energy (Hamiltonian); KE + PE

3. Try to minimize it w.r.t the parameters. $\frac{\partial E}{\partial \theta _i} = 0$

4. Finally obtain the energy with the obtained parameter(s).

Example:

Consider the Hamiltonian for the 1D harmonic oscillator:

$$H = - \frac{-\hbar ^2}{2m}\, \frac{d^2}{dx^2} + \frac{1}{2} mw^2 x^2\, \ldots \, (5)$$

If we want to find an approximation or an upper bound for the ground state energy, say we start with a trial function of Gaussian form $Ae^{-bx^2}$.

First we can use the normalization criterion to get $A$ in terms of $b$.

$$\int \psi ^* \psi \, dx = |A|^2 \int \limits_{-\infty}^{+\infty} e^{-2bx^2} \, dx = |A|^2 \, \sqrt{\frac{\pi}{2b}} = 1; \, A = \left(\frac{2b}{\pi}\right)^{\frac{1}{4}}\, \ldots \, (6)$$


Let's calculate the K.E.

$$\langle T \rangle = -\frac{\hbar ^2}{2m}\, |A|^2 \, \int \limits_{-\infty}^{+\infty} e^{-bx^2}\, \frac{d^2}{dx^2}\left(e^{-bx^2}\right)\, dx = \frac{\hbar ^2 b}{2m}\, \ldots \, (7)$$  

Similarly for P.E.

$$\langle V \rangle = -\frac{1}{2}\, m w^2 |A|^2 \, \int \limits_{-\infty}^{+\infty} e^{-2bx^2}\, x^2\, dx = \frac{mw^2}{8b}\, \ldots \, (8)$$

Let's minimize the Hamiltonian w.r.t the parameter $b$

$$\frac{\partial H}{\partial b} = \frac{\partial }{\partial b}\left(\frac{\hbar ^2b}{2m} + \frac{mw^2}{8b}\right) = \frac{\hbar ^2}{2m} - \frac{mw^2}{8b^2} = 0 \, \implies b = \frac{mw}{2\hbar}\, \ldots \, (9)$$

Putting this $b$ back to the Hamiltonian, we get: $H = \frac{1}{2} \hbar w$.


This is indeed the ground state energy of the 1d harmonic oscillator.

--------------------------------------------------

In the above example we're lucky to land in the exact ground state but this may not be the case always.

What about the delta function potential; The solution for this is also known but let's try finding the ground state using Variational principle; The Hamiltonian is given by

$$H = -\frac{-\hbar ^2}{2m}\, \frac{d^2}{dx^2} - \alpha\delta (x)$$

the minus signifies attractive delta function potential.

Just as before our trial wave function is Gaussian $Ae^{-bx^2}$ and we know the normalization constant is $A=\left(\frac{2b}{\pi}\right)^{\frac{1}{4}}$.

We also know the expectation for the K.E. from before. But the potential energy would be different

$$V = -\alpha \,\int \limits_{\infty}^{\infty} e^{-2bx^2}\, \delta(x)\, dx = -\alpha \sqrt{\frac{2b}{\pi}}$$

We proceed as before

$$\frac{\partial H}{\partial b} = \frac{\partial }{\partial b}\left(\frac{\hbar ^2b}{2m} - \alpha \sqrt{\frac{2b}{\pi}}\right) = \frac{\hbar ^2}{2m} - \frac{\alpha}{\sqrt{2b\pi}} = 0 \, \implies b = \frac{2m^2\alpha ^2}{\pi \hbar ^4}$$

Putting this together we get the ground state energy as $H_0 = -\frac{m\alpha ^2}{\pi \hbar ^2}$

This is indeed pretty close to the true ground state energy $\left(-\frac{m\alpha ^2}{2\hbar ^2}\right)$  but slightly higher.

### Schrodinger Equation for Multi-Electron Atom

Once we move beyond Hydrogen atom which has a proton and an electron, we enter the realm of multi-electron atom, say Helium atom $\left({}^2\text{He}_4 \right)$ with 2 protons and 2 electrons. Solving the Schrodinger's equation like for the Hydrogen atom becomes impossible because of the introduction of an interaction term between electrons;

Let's see this; we will have K.E terms for individual electrons, potential energy terms for each electron (interaction with positively charged nucleus), and potential energy term for the mutual repulsion of each pair of negatively charged electrons:


$$H = -\frac{ℏ^2}{2m_e}(∇^2_1+∇^2_2) + V(r_1) + V(r_2) + V(r_{12})\, \ldots \, (10)$$

where

$$V(r_1) = \frac{-2e^2}{4\pi ϵ_0 r_1},\,  V(r_2) = \frac{-2e^2}{4\pi ϵ_0 r_2}, \, V(r_{12}) = \frac{e^2}{4\pi ϵ_0 |r_1-r_2|}\, \ldots \, (11)$$

From this we can write the generalized expressions for Hamiltonian for any atoms with atomic number $Z$

$$H = -\frac{ℏ^2}{2 m_e} \sum \limits _i ∇^2_i + \sum \limits _i V(r_i) + \sum \limits_{i, \, j\\ i\neq j} V(r_{ij})\, \ldots \, (12)$$

Because of the interaction term when we want to solve the Schrodinger Equation $H|\psi⟩ = E|\psi\rangle$ it becomes impossible to separate it for individual electrons; Neglecting the repulsion term we can write:

$$H = -\frac{ℏ^2}{2m_e}(∇^2_1+∇^2_2) + V(r_1) + V(r_2) → H(r_1) + H(r_2)\, \ldots \, (13)$$

so, we end up with two independent Schrödinger equations that are exactly the same as that for the hydrogen atom, except that the nuclear charge is $2e$ rather than $e$.

$$\begin{align}H(r_1)φ(r_1) &= E_1φ(r_1)\\ H(r_2)φ(r_2) &= E_2φ(r_2) \end{align} \, \ldots \, (14)$$

The fact that electrons interact through their electron-electron repulsion  means that an exact wavefunction for a multi-electron system would be a _single function that depends simultaneously upon the coordinates of all the electrons_; i.e., a multi-electron wavefunction: $|ψ(r_1, r_2, \ldots r_i)⟩$


### Hartree Approximation

The separable Hamiltonians for electrons (e.g. in Helium) once we neglect the mutual repulsion between them, tell us that the total electronic wave function would just be the product of the two Hydrogen atom wavefunctions

$$|ψ(r_1, r_2)⟩ = |ψ(r_1)\rangle \, |ψ(r_2)\rangle \, \ldots \, (15)$$

Neglecting the repulsion between electrons has a serious consequence but the Hartree approximation starts by invoking an initial ansatz that the multi-electron wavefunction can be expanded as a product of single electron wavefunctions

$$|ψ(r_1, r_2, \ldots r_i)⟩ \approx |ψ(r_1)\rangle \, |ψ(r_2)\rangle \ldots |ψ(r_i)\rangle \, \ldots \, (16)$$

With this one can write the one electron Schrodinger equations:

$$-\frac{ℏ^2}{2m_e} \, ∇^2\psi _i(r) + V(r)\, \psi_i = \epsilon _i (r) \psi _i(r)\, \ldots \, (17)$$

The potential term $V(r)$ consists of nuclear electron interaction and the mean field arising from $N-1$ electron.

The K.E. would be

$$\left\langle\phi_j\left|\frac{ℏ^2}{2m_e} \, ∇^2\right|\phi_j\right\rangle$$

Nuclear-electron energy (potential)

$$\left\langle\phi_j\left|\frac{-Ze^2}{|r-R|} \right|\phi_j\right\rangle$$

Electron-electron repulsive energy (potential)

$$\left\langle\phi_j(r)\, \phi_k(r')\left|\frac{e^2}{|r-r'|} \right|\phi_j(r)\phi _k(r')\right\rangle$$


Adding these all together to get the total energy $ϵ_j$ of the orbital $ϕ_j$

$$\epsilon _j = \left\langle\phi_j\left|\frac{ℏ^2}{2m_e} \, ∇^2\right|\phi_j\right\rangle + \left\langle\phi_j\left|\frac{-Ze^2}{|r-R|} \right|\phi_j\right\rangle + \left\langle\phi_j(r)\, \phi_k(r')\left|\frac{e^2}{|r-r'|} \right|\phi_j(r)\phi _k(r')\right\rangle \, \ldots \, (18)$$
