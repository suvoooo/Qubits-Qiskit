### Chapter 2: Entanglement in Qubits

### Entanglement 

Two quantum systems are said to be entangled when the values of certain properties of one system is non-classically correlated with the values the same properties will assume for other system. 

Let's get started with the one of the simplest examples of entangled states, known as Bell State. The general form of Bell State: 

$|\beta _{xy}\rangle = \frac{|0y\rangle + (-1)^x\, |1\bar{y}\rangle}{\sqrt{2}}$. 

The $x$ and $y$ are called  phase bit and parity beat. Consider one of the Bell states : $|\beta _{01}\rangle = \frac{|01\rangle + |10\rangle}{\sqrt{2}}$,  similarly we can write the others..  

--------------------------------------------

**_How to simply tell when a state is entangled?_**

--------------------------------

When two states are not entangled, they are said to be in separable or product state. Suppose that $H_1, H_2$ are 2 Hilbert spaces with dimensions $N_1, N_2$ respectively. We can always put these Hilbert spaces together using tensor product to construct a larger Hilbert space. This larger space if denoted by $H = H_1\otimes H_2$ will have dimension $N=N_1\times N_2$. Let's consider a state $|\psi \rangle$ from Hilbert space $H_1$ and $|\phi \rangle$ from Hilbert space $H_2$. If a new state $|\chi\rangle$ can be formed as $|\chi \rangle = |\psi \rangle \otimes |\phi \rangle$ then $|\chi \rangle$ is a separable (also known as product state) state.  

The simplest test that can be performed for states in $\mathbb{C}^4$ is when $|\chi \rangle$ is written in column vector form as -- $$ \begin {pmatrix} a\\b \\c \\d \end{pmatrix}, $$ and if $ad=bc$ then the state is separable. 

Let's test this with 2 of the Bell states: 
$|\beta _{01}\rangle = \frac{|01\rangle + |10\rangle}{\sqrt{2}}, \, |\beta _{10}\rangle = \frac{|00\rangle - |11\rangle}{\sqrt{2}}$, when written as column vector  with basis vectors as $\{|00\rangle , |01\rangle, |10\rangle, |11\rangle\}$, takes the form: 
$$|\beta _{01}\rangle =  \frac{1}{\sqrt{2}}\begin {pmatrix} 0\\1 \\1 \\0 \end{pmatrix}\, \text{and}\, |\beta _{10}\rangle = \begin {pmatrix} 1\\0 \\0 \\1 \end{pmatrix}\, .$$
Clearly none of them satisfy $ad = bc$, so these are entangled state. Great ! 

The 4 Bell states are maximally entangled pure states, pure not in the $\{|0\rangle , |1\rangle\}$ basis but pure in $\mathbb{C} ^4$ basis.

### Spooky Action at a Distance ?  

-------------------------------------

Let's take one of the Bell state for example and understand what is entanglement. We consider two observers Bob and Alice and rewrite the Bell State $|\beta _{11}\rangle$ in a different format. 

$|\beta _{11}\rangle = \frac{|0\rangle_{A}\otimes |1\rangle_{B}- |1\rangle_A \otimes |0\rangle_B}{\sqrt{2}}$. 

What the above expression tells us is that if Alice makes a measurement in the $\{|0\rangle, |1\rangle\}$ eigen basis of Alice, there are two outcomes. If Alice measures $|0\rangle$ the state of the second particle must be $|1\rangle$ (measured by Bob). We say that the system has collapsed in $|0\rangle_A|1\rangle _B$ state. Or, if Alice measures $|1\rangle$ the state of the second particle must be $|0\rangle$ (measured by Bob). The system has collapsed in $|1\rangle_A|0\rangle _B$ state. 

The measurement made by Alice could be perfectly random, either possibility of $|0\rangle\,  \text{or}\, |1\rangle$ having probability $\frac{1}{2}$. But when Bob measures the qubit it will be opposite of what Alice measured. Individually for Bob it will be random outcome at first sight but if Alice and Bob communicates then they would figure out that the outcomes are completely correlated.  Even when they are separated by a large distance the state of qubit cannot be described independently and this quite a fundamental difference from classical physics. Thus choice of measurement at one location is affecting the appears to affect the system at another location. This phenomenon was termed by _'Spooky action at a distance'_ by Einstein. 

Based on this Einstein, Podolsky and Rosen proposed a thought experiment in the paper published in [Physical Review](http://www.drchinese.com/David/EPR.pdf)  and argued that the physical reality provided by quantum theory is imcomplete and, whether it is possible to construct a theory that fulfill the 'local realism'.  


Bell however showed in his [paper](https://cds.cern.ch/record/111654/files/vol1p195-200_001.pdf)  that using simple probability theory it is possible to prove that the correlations between the qubits can be predicted by quantum mechanics, no new theory or hidden variable is necessary to explain the non-local nature of entanglement.

#### How to Create a Bell State Using Simple Circuit

------------------------------

Once we have gone through the background and little bit of necessary mathematics that will be required later on related to entanglement, now it is time to construct simple quantum circuits that will create Bell state(s). 

Before we have said that $\{|0\rangle ,\, |1\rangle \}$ forms an orthonormal basis. Let's introduce another orthnormal basis set for $\mathbb{C}^2$. We denote this vectors by $$|+\rangle = \frac{1}{\sqrt 2}\begin{pmatrix} 1\\1 \end{pmatrix}, |-\rangle = \frac{1}{\sqrt 2}\begin{pmatrix} 1\\-1 \end{pmatrix}.$$ The Unitary matrix necessary for a transformation from $\{|0\rangle, |1\rangle \}$ to $\{|+\rangle, |-\rangle \}$ (for the transformation rule check the previous [notebook](https://github.com/suvoooo/Qubits-Qiskit/blob/master/QCwithQskit-1.ipynb)) is below-- 

$$\frac{1}{\sqrt 2}\begin{pmatrix} 1& 1 \\ 1 & -1\end{pmatrix}. $$
This matrix is also known as **Hadamard Matrix** and we will use this later to construct Bell State. 

Let's spend a bit more time with Hadamard operator because action of Hadamard operator on the standard basis $\{|0\rangle, \, |1\rangle \}$ will create the superposition state. $H|0\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}}$ and $H|1\rangle = \frac{|0\rangle - |1\rangle}{\sqrt{2}}$. In a more general form the Hadamard operator change the state of the qubit from $|\psi \rangle = \alpha |0\rangle + \beta |1\rangle$ to $|\psi \rangle = \frac{\alpha + \beta}{\sqrt 2} |0\rangle + \frac{\alpha - \beta}{\sqrt 2} |1\rangle$. This also implies that after the Hadamard operation probability of finding qubit in state $|0\rangle$ is now $\left|\frac{\alpha + \beta}{2}\right|^2$.  

Now, we can think step by step how to create a Bell state using the simple circuits we have leared so far. 

Let's write down the 4 Bell states explicitly, 

$|\beta _{00}\rangle = \frac{|00\rangle + |11\rangle}{\sqrt{2}}, \, |\beta _{01}\rangle = \frac{|01\rangle + |10\rangle}{\sqrt{2}}, \, |\beta _{10}\rangle = \frac{|00\rangle - |11\rangle}{\sqrt{2}}, |\beta _{11}\rangle = \frac{|01\rangle - |10\rangle}{\sqrt{2}}$.  

We see there are 2 qubits involved and one of the simplest 2-qubit gate we know is CNOT (CX). We discussed about CX gate and corresponding operator in [last notebook](https://github.com/suvoooo/Qubits-Qiskit/blob/master/QCwithQskit-1.ipynb). Let's write the CX gate operator in outer product form --
$CX = |00\rangle \langle 00| + |01\rangle \langle 01| + |10\rangle \langle 11| + |11\rangle \langle 10|$.

Let's consider a thought expreiment where the control qubit is in superposition state, let's say $|c\rangle = |0\rangle + |1\rangle$ and the target qubit is in state $|0\rangle$. So, if we apply CNOT gate on this system $\left(|00\rangle + |10\rangle \right)$ we get :

$CX\left(|00\rangle + |10\rangle \right) = \left(|00\rangle \langle 00| + |01\rangle \langle 01| + |10\rangle \langle 11| + |11\rangle \langle 10| \right) (|00\rangle + |10\rangle)$ 
$=|00\rangle + |11\rangle $. 

This is already one of the Bell states $(\beta _{00})$ times $\frac{1}{\sqrt{2}}$. Now, let's look into the action of the Hadamard Gate on the qubit $|0\rangle$, which creates $\frac{|0\rangle + |1\rangle}{\sqrt{2}}$. So that's it ! If we start with Hadamard gate and create the superposition state and this act as control qubit for the CNOT gate afterwards then we can create entangled qubits. Let's get started with the circuits. 



```
# # install and import qiskit 
# !pip3 install qiskit
```

```
import qiskit as q
import matplotlib.pyplot as plt
from qiskit import IBMQ
```

```

IBMQ.save_account('putyourtoken')
```

```
IBMQ.load_account()
```

```
qr = q.QuantumRegister(2)
cr = q.ClassicalRegister(2)

circuit = q.QuantumCircuit(qr, cr)
circuit.draw()
```

```
# apply the Hadamard gate on 0th qubit
circuit.h(qr[0])
circuit.draw(output='mpl') 
```

```
# now we apply the cnot gate 
# as explained before the control qubit should be in superposition state 
circuit.cx(qr[0], qr[1])
circuit.draw(output='mpl')
```

```
circuit.measure(qr, cr) # measure quantum bit into classical bit
circuit.draw(output='mpl')
```

#### Simulate the Result on a Local Computer 

We are preparing the Bell state $\beta _{00} = \frac{|00\rangle + |11\rangle}{\sqrt{2}}$, as we can see that there's $50\%$ probability of measuring the state in $|00\rangle$ and $50\%$ chance in $|11\rangle$ state. Also there is $0\%$ chance of measuring $|01\rangle$ and $|10\rangle$ states. On a local simulator we expect $100\%$ accurate measurement.   

```
# let's simulate the results on a local computer

simulator = q.Aer.get_backend('qasm_simulator') # qasm : quantum assembly lang.
results = q.execute(circuit, backend=simulator).result() # execute the circuit based on the simulator as backend
q.visualization.plot_histogram(results.get_counts(circuit))
```

As expected, we get almost $50\%$ chance for each of $|00\rangle, \, |11\rangle$ states. 

Let's now run the same circuit in a quantum computer. We use IBM Quantum Computer, see below-- 
 

```
provider = IBMQ.get_provider('ibm-q')
qcomp = provider.get_backend('ibmq_london') # check which computer has 0 jobs on queue
job = q.execute(circuit, backend=qcomp)
q_result = job.result()

q.visualization.plot_histogram(q_result.get_counts(circuit))
```

So here we see that $|00\rangle, \, |11\rangle$ states have almost $50\%$ probabilities, but $\left(|10\rangle, |01\rangle \right)$ states which according to the theory should not be present but, due to unavoidable quantum noises, they are present too.  

The combined state of the qubits cannot be written as 2 seperate qubits. As discussed before measuring one qubit at state $|0\rangle$ will make the collective system to collapse the state to $|00\rangle$. This we discussed earlier as spooky action in distance.
